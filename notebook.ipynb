# Ex1

```python
# Soit une population des empolyés caractérisée par leurs salaires et leurs âges
# Soient deux employés Ali (salaire=1600 et age=38) et Samir (salaire=2100 et age=43)
# Soit une requete (salaire=1500 et age=35)
# Quel est l'employé le plus proche de la requete ?
# On peut utiliser la distance euclidienne pour calculer la distance entre deux employés

import numpy as np
#qimport pandas as pd
import matplotlib.pyplot as plt
```

## Representation mathématique des données

```python

Ali = np.array([1600, 38])
Samir = np.array([2100, 43])
requete = np.array([1500, 35])
```

## Calcul Distance

```python

# Calculer la distance euclidienne entre Ali et la requete
# Calculer la distance euclidienne entre Samir et la requete
# Quel est l'employé le plus proche de la requete ?

# Solution
Ali_requete = np.linalg.norm(Ali - requete)
Samir_requete = np.linalg.norm(Samir - requete)
print(Ali_requete, Samir_requete)
```

## Visualisation

```python

# Visualisation
plt.scatter(Ali[0], Ali[1], color='red', label='Ali')
plt.scatter(Samir[0], Samir[1], color='blue', label='Samir')
plt.scatter(requete[0], requete[1], color='green', label='requete')
plt.legend()
plt.show()
```

# Ex2

```python
# New data
Ali = np.array([800.5, 25])
Sonia = np.array([1500.65,40])
Rahma = np.array([1800.77,43])
Salah = np.array([2100.566,32])
Walid = np.array([1400.56,51])
# make a matrix of the data
data = np.array([Ali, Sonia, Rahma, Salah, Walid])
names = np.array(['Ali', 'Sonia', 'Rahma', 'Salah', 'Walid'])
request = np.array([2000, 40])
```

```python
data
```

```python
# calculate euclidien distance between request and each employee and store it in a vector
distances = np.linalg.norm(data - request, axis=1)
```

```python
distances
```

```python
order=np.argsort(distances)
```

```python
order
```

```python
# order the names according to the order of the distances
names[order]
```

```python
# Visualisation
# Show each employee with a different color
# Show the request with a different color

plt.scatter(Ali[0], Ali[1], color='red', label='Ali')
plt.scatter(Sonia[0], Sonia[1], color='blue', label='Sonia')
plt.scatter(Rahma[0], Rahma[1], color='green', label='Rahma')
plt.scatter(Salah[0], Salah[1], color='yellow', label='Salah')
plt.scatter(Walid[0], Walid[1], color='black', label='Walid')
plt.scatter(request[0], request[1], color='pink', label='request')
plt.legend()
plt.show()
```

# Probabilité & Statistiques

```python
from numpy.random import standard_normal, normal, multivariate_normal
```

```python
# Generate random data following a normal distribution with mean=0 and std=1 (standard normal distribution)
standard_normal(5)
```

```python
x=normal(30,0.00001,5)
```

```python
np.mean(x)
```

```python
# mu: mean of the distribution (vector)
mu=np.array([1500,30])
# sigma: covariance matrix (matrix)
sigma=np.array([[0.00001,0],[0,0.00001]])
# Generate random data following a normal distribution with mean=mu and covariance=sigma
multivariate_normal(mean=mu, cov=sigma, size=5)
```
